```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [48,47,48,49,50,51,52]
```

```python
plt.plot(x, y, 'r-', markersize = 30, linewidth = 5)
```

```python
plt.xlabel("week")
plt.ylabel("tempreture")
plt.title("weather")
plt.plot(x, y, color = 'green')
```

```python
plt.xlabel("week")
plt.ylabel("tempreture")
plt.title("weather")
plt.plot(x, y, 'r+', markersize = 50, markerfacecolor = 'black')
```

```python
plt.plot(x, y, 'r-', alpha = 0.9)
```
