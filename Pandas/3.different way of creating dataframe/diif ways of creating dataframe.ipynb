```python
import pandas as pd
```

```python
#1st way to create dataframe
df = pd.read_csv('weather_data.csv')
df
```

```python
#2nd way to create dataframe
df = pd.read_excel('weather.xlsx','Sheet1')
df
```

```python
#3rd way to create dataframe
weather = {
    'day' : ['1/1/2017', '1/2/2017', '1/3/2017','1/4/2017'],
    'temperature' : ['32', '24', '24', '25'],
    'windspeed' : ['2', '3', '4', '5'],
    'event' : ['rain', 'sunny', 'snow', 'rain']
}
df = pd.DataFrame(weather)   
df         
```

```python
#4th way to create dataframe
weather = [
     ['1/1/2017', '32', '2', 'rain'],
     ['1/2/2017', '24', '3', 'sunny'],
     ['1/3/2017', '24', '4', 'snow'],
     ['1/4/2017', '25', '5', 'rain']
]
df = pd.DataFrame(weather, columns = ['days', 'temperature', 'windspeed', 'event'])
df
```

```python
#5th way to create dataframe
weather = [
    {'days' : '1/1/2020', 'temperature' : '32', 'windspeed' : '2', 'event' : 'rain'},
    {'days' : '1/2/2020', 'temperature' : '24', 'windspeed' : '3', 'event' : 'sunny'},
    {'days' : '1/3/2020', 'temperature' : '24', 'windspeed' : '4', 'event' : 'snow'},
    {'days' : '1/4/2020', 'temperature' : '25', 'windspeed' : '5', 'event' : 'rain'}
]
df = pd.DataFrame(weather)
df
```
