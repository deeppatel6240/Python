```python
def convert_windspeed_cell(cells):
    if cells == 'na':
        return '0';
    return cells;

def convert_windspeeds_cell(cells):
    if cells == '-4':
        return '00';
    return cells;

df2 = pd.read_excel('new2.xlsx', 'Sheet1', converters = {'windspeed' : convert_windspeed_cell,
                                                         'windspeed' : convert_windspeeds_cell}, inplace =True)
df2
```

```python
df2.to_excel('new3.xlsx', index =False)
```
