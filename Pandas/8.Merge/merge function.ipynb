```python
import pandas as pd
```

```python
Sub_Marks = {
    'Subject' : ['Maths', 'C', 'Python'],
    'Deep' : ['78', '87', '89'],
    'Raj' : ['85', '25', '65'],
    'Hardik' : ['72','15', '90']
}
df1 = pd.DataFrame(Sub_Marks)
df1
```

```python
Sub_Marks1 = {
    'Subject' : ['Maths', 'C', 'Python'],
    'Deep' : ['65', '87', '81'],
    'Raj' : ['85', '45', '62'],
    'Hardik' : ['73','45', '87']
}
df2 = pd.DataFrame(Sub_Marks1)
df2
```

```python
df3 = pd.merge(df1,df2, on= 'Subject')
df3
```

```python
Sub_Marks2 = {
    'Subject' : ['Maths', 'C', 'Python', 'c++'],
    'Deep' : ['78', '87', '89', '54'],
    'Raj' : ['85', '25', '65', '48'],
    'Hardik' : ['72','15', '90', '98']
}
df3 = pd.DataFrame(Sub_Marks2)
df3
```

```python
df4 = pd.merge(df1,df2, on= 'Subject', how = 'outer')
df4
```

```python
df4 = pd.merge(df1,df2, on= 'Subject', how = 'inner')
df4
```

```python
df4 = pd.merge(df1,df2, on= 'Subject', how = 'left')
df4
```

```python
df4 = pd.merge(df1,df2, on= 'Subject', how = 'right')
df4
```

```python
df5 = pd.merge(df1,df2,on="Subject",how="outer",indicator=True)
df5
```

```python
df6 = pd.merge(df1,df2,on="Subject",how="outer",suffixes = ['_first','_Second'])
df6
```
