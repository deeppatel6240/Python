```python
import pandas as pd
```

```python
df =pd.read_csv('Sub_Marks.csv', parse_dates = ['Dates'])
df
```

```python
type(df.Dates[0])
```

```python
df.set_index('Dates', inplace = True)
```

```python
df
```

```python
#1st method to fill missing data
df1 = df.fillna(1)
df1
```

```python
#diff cols have diff values for NaN
df2 = df.fillna(
    {'Grades' : 'Very bad',
    'Deep' : '50',
    'Raj' : '60',
    'Hardik' : '70',
    'Dhruv' : '80',
    'Kishan' : '90'}    
)
```

```python
df2
```

```python
#forward fill 
df3 = df.fillna(method = 'ffill')
df3
```

```python
#backward fill
df4 = df.fillna(method = 'bfill')
df4
```

```python
#interpolate method
df5 = df.interpolate()
df5
```

```python
#drop all rows which have a NaN values
df6 = df.dropna()
df6
```
